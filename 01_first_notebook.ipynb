```python
import pandas as pd 
import ipywidgets as widgets
from IPython.display import display, clear_output

# Global dataframe holder
loaded_df = None

# UI elemnts
file_path_input = widgets.Text(
    value="orders.csv",
    placeholder="Enter CSV file path",
    description="CSV Path:",
    layout = widgets.Layout(width="600px")
)

load_button = widgets.Button(
    description="Load CSV",
    button_style="primary"
)

show_all_table_button = widgets.Button(
    description = "Show All",
    button_style = "info"
)

show_head_button = widgets.Button(
    description = "Show Head",
    button_style="info"
)

show_shape_button = widgets.Button(

    description="Show Shape",
    button_style="info"
)

show_columns_button = widgets.Button(

    description="Show Columns",
    button_style="info"
)

show_dtypes_button = widgets.Button(

    description="Show Data Types",
    button_style="info"
)

output = widgets.Output(
    layout=widgets.Layout(
        border="1px solid #ddd",
        padding="10px",
        margin="10px 0px 0px 0px"
    )
)

def check_data_loaded():
    if loaded_df is None:
        print("Please load a CSV file first")
        return False
    return True

# Button click function
def load_csv_button_clicked(button):
    global loaded_df
    
    with output:
        clear_output()

        file_path = file_path_input.value.strip()
        try:
            loaded_df = pd.read_csv(file_path)

            print("CSV loaded successfully")
            print(f"Rows: {loaded_df.shape[0]}")
            print(f"Columns: {loaded_df.shape[1]}")

            display(loaded_df.head())
            
        except Exception as error:
            print("Error while loading CSV file : ")
            print(error)

def show_table_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        print("Showing all rows and columns")
        display(loaded_df)

def show_head_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return
        
        print("First 5 rows:")
        display(loaded_df.head())

def show_shape_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        rows, columns = loaded_df.shape

        print("Dataset shape:")
        print(f"Rows: {rows}")
        print(f"Columns: {columns}")

def show_columns_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        print("Columns:")
        for column in loaded_df.columns:
            print(f" - {column}")

def show_dtypes_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return
            
        print("Data types:")
        display(loaded_df.dtypes.to_frame(name="data_type"))

# connect button to function
load_button.on_click(load_csv_button_clicked)
show_all_table_button.on_click(show_table_clicked)
show_head_button.on_click(show_head_clicked)
show_shape_button.on_click(show_shape_clicked)
show_columns_button.on_click(show_columns_clicked)
show_dtypes_button.on_click(show_dtypes_clicked)

# Display UI
app_ui = widgets.VBox([
    widgets.HTML("<h1>Mini Data Analysis Toolkit</h1>"),
    widgets.HTML("<p>v0.2 - CSV Loader and Data Preview</p>"),
    file_path_input,
    widgets.HBox([
        load_button
    ]),
    widgets.HBox([
        show_all_table_button,
        show_head_button,
        show_shape_button,
        show_columns_button,
        show_dtypes_button
    ]),
    output
])

display(app_ui)
```
