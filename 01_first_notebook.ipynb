```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

orginal_df = None


drop = widgets.Dropdown(
    options=["Please selected file tpye", 'CSV', 'Excel', 'API'],
    value='Please selected file tpye',
    description='',
    disabled=False,
)

drop.style.description_width = "0px"

file_path_input = widgets.Text(
    value = "/Users/oguzhan/Desktop/code/oguzhanmst/data-analysis-toolkit/e_ticaret_banka_islemleri_kirli_veri.csv",
    placeholder="Enter CSV file path",
    description="CSV Path:",
    layout=widgets.Layout(width="100%")
)

load_button = widgets.Button(

    description="Load Data",
    button_style="primary"

)

info_output = widgets.Output()
original_output = widgets.Output()
edit_output = widgets.Output()
analysis_output = widgets.Output()
message_output = widgets.Output()

def check_drop_input():
    if drop.value == "CSV" and file_path_input.value.strip() == "":
        print("please")
        return False
    return True
    
def show_table(button):
    global orginal_df

    with info_output:
        clear_output()
        
        if not check_drop_input():
            return

        file_type = drop.value
        file_path = file_path_input.value.strip()

        if file_type == "CSV":
            orginal_df = pd.read_csv(file_path)
        else:
            print("hata")
            
        print("successfuly")
        display(orginal_df)

tabs = widgets.Tab(
    children=[
        info_output,
        original_output,
        edit_output,
        analysis_output
    ]
)

tabs.set_title(0, "Info")
tabs.set_title(1, "Table")
tabs.set_title(2, "Edit Table")
tabs.set_title(3, "Analysis")

load_button.on_click(show_table)

app_ui = widgets.VBox([

    widgets.HTML("""<h1 style="margin: 5px 0 0 0; padding: 0;">Mini Data Analysis Toolkit</h1>"""),
    widgets.HTML("""<p style="margin: 0 0 5px 0; padding: 0;">Tab Based UI Structure</p>"""),
    widgets.HBox([
        drop,
        file_path_input,
        load_button
    ]),
    tabs
])

display(app_ui)
```
