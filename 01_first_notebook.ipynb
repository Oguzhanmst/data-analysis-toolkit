```python
import pandas as pd 
import ipywidgets as widgets
from IPython.display import display, clear_output

# Global dataframe holder
loaded_df = None

#Global copy dataframe holder
copy_df = None

# UI elemnts
file_path_input = widgets.Text(
    value="orders.csv",
    placeholder="Enter CSV file path",
    description="CSV Path:",
    layout = widgets.Layout(width="600px")
)

load_button = widgets.Button(
    description="Load CSV",
    button_style="primary"
)

show_all_table_button = widgets.Button(
    description = "Show All",
    button_style = "info"
)

show_head_button = widgets.Button(
    description = "Show Head",
    button_style="info"
)

show_shape_button = widgets.Button(

    description="Show Shape",
    button_style="info"
)

show_columns_button = widgets.Button(

    description="Show Columns",
    button_style="info"
)

show_dtypes_button = widgets.Button(

    description="Show Data Types",
    button_style="info"
)

show_missing_button = widgets.Button(
    description="Show Missing",
    button_style="warning"
)

show_duplicates_button = widgets.Button(
    description="Show Duplicates",
    button_style="warning"
)

show_summary_button = widgets.Button(
    description="Show Summary",
    button_style="success"
)

drop_duplicates_button = widgets.Button(
    description="Drop Duplicates",
    button_style="danger"
)

drop_missing_rows_button = widgets.Button(
    description="Drop Missing Rows",
    button_style="danger"
)

reset_data_button = widgets.Button(
    description="Reset Data",
    button_style="warning"
)

output = widgets.Output(
    layout=widgets.Layout(
        border="1px solid #ddd",
        padding="10px",
        margin="10px 0px 0px 0px"
    )
)

#Is the data uploaded
def check_data_loaded():
    if copy_df is None:
        print("Please load a CSV file first")
        return False
    return True

# Button click function
def load_csv_button_clicked(button):
    global loaded_df, copy_df
    
    with output:
        clear_output()

        file_path = file_path_input.value.strip()

        #Path check 
        if file_path == "":
            print("Please enter a CSV file path.")
            return
            
        #Path check end
        if not file_path.endswith(".csv"):
            print("Please enter a valid CSV file.")
            return
        
        try:
            loaded_df = pd.read_csv(file_path)
            copy_df = loaded_df.copy()

            print("CSV loaded successfully")
            print(f"Rows: {loaded_df.shape[0]}")
            print(f"Columns: {loaded_df.shape[1]}")

            display(loaded_df)
            
        except Exception as error:
            print("Error while loading CSV file : ")
            print(type(error).__name__)
            print(error)

#Button show the all table function
def show_table_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        print("Showing all rows and columns")
        display(copy_df)

#Button show the first five rows function
def show_head_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return
        
        print("First 5 rows:")
        display(copy_df.head())

#Button show the number of columns and rows function
def show_shape_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        rows, columns = copy_df.shape

        print("Dataset shape:")
        print(f"Rows: {rows}")
        print(f"Columns: {columns}")

#Button show the columns names of the table function
def show_columns_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        print("Columns:")
        for column in copy_df.columns:
            print(f" - {column}")

#Button show the columns types 
def show_dtypes_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return
            
        print("Data types:")
        display(copy_df.dtypes.to_frame(name="data_type"))

#Button show missing values function
def show_missing_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        print("Missing values by column:")

        missing_values = copy_df.isnull().sum().to_frame(name="missing_count")
        missing_values["missing_percantage"] = (
            copy_df.isnull().mean() * 100 
        ).round(2)

        display(missing_values)

# Button show duplicate row count function
def show_duplicates_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return
            
        duplicate_count = copy_df.duplicated().sum()
        print("Duplicate row count:")
        print(duplicate_count)

# Button show numerical summary function
def show_summary_clicked(button):
    with output:
        clear_output()

        if not check_data_loaded():
            return

        print("Numerical summary:")
        display(copy_df.describe())

def drop_duplicates_clicked(button):
    global copy_df

    with output:
        clear_output()

        if not check_data_loaded():
            return

        before_rows = copy_df.shape[0]
        copy_df = copy_df.drop_duplicates()
        after_rows = copy_df.shape[0]

        removed_rows = before_rows - after_rows

        print("Duplicate rows removed.")
        print(f"Rows before: {before_rows}")
        print(f"Rows after: {after_rows}")
        print(f"Removed rows: {removed_rows}")

        display(copy_df.head())

def drop_missing_rows_clicked(button):
    global copy_df

    with output:
        clear_output()

        if not check_data_loaded():
            return

        before_rows = copy_df.shape[0]
        copy_df = copy_df.dropna()
        after_rows = copy_df.shape[0]

        removed_rows = before_rows - after_rows

        print("Rows with missing values removed.")
        print(f"Rows before: {before_rows}")
        print(f"Rows after: {after_rows}")
        print(f"Removed rows: {removed_rows}")

        display(loaded_df)

def reset_data_clicked(button):
    global copy_df

    with output:
        clear_output()

        if loaded_df is None:
            print("Please load a CSV file first.")
            return

        copy_df = loaded_df.copy()

        print("Data has been reset to the original loaded version.")
        print(f"Rows: {copy_df.shape[0]}")
        print(f"Columns: {copy_df.shape[1]}")

        display(copy_df)

# connect button to function
load_button.on_click(load_csv_button_clicked)
show_all_table_button.on_click(show_table_clicked)
show_head_button.on_click(show_head_clicked)
show_shape_button.on_click(show_shape_clicked)
show_columns_button.on_click(show_columns_clicked)
show_dtypes_button.on_click(show_dtypes_clicked)
show_missing_button.on_click(show_missing_clicked)
show_duplicates_button.on_click(show_duplicates_clicked)
show_summary_button.on_click(show_summary_clicked)
drop_duplicates_button.on_click(drop_duplicates_clicked)
drop_missing_rows_button.on_click(drop_missing_rows_clicked)
reset_data_button.on_click(reset_data_clicked)

# Display UI
app_ui = widgets.VBox([
    widgets.HTML("<h1>Mini Data Analysis Toolkit</h1>"),
    widgets.HTML("<p>v0.2 - CSV Loader and Data Preview</p>"),
    file_path_input,
    widgets.HBox([
        load_button
    ]),
    widgets.HBox([
        show_all_table_button,
        show_head_button,
        show_shape_button,
        show_columns_button,
        show_dtypes_button
    ]),
    widgets.HBox([
    show_missing_button,
    show_duplicates_button,
    show_summary_button
]),
    widgets.HBox([
    drop_duplicates_button,
    drop_missing_rows_button,
    reset_data_button
]),
    output
])

display(app_ui)
```
