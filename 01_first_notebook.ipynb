```python
import pandas as pd 
import ipywidgets as widgets
from IPython.display import display, clear_output

# Global dataframe holder
loaded_df = None

# UI elemnts
file_path_input = widgets.Text(
    value="orders.csv",
    placeholer="Enter CSV file path",
    description="CSV Path:",
    layout = widgets.Layout(width="600px")
)

load_button = widgets.Button(
    description="Load CSV",
    button_style="primary"
)

output = widgets.Output(
    layout=widgets.Layout(
        border="1px solid #ddd",
        padding="10px",
        margin="10px 0px 0px 0px"
    )
)

# Button click function
def load_csv_button_clicked(button):
    global loaded_df
    
    with output:
        clear_output()

        file_path = file_path_input.value.strip()
        try:
            df = pd.read_csv(file_path)

            print("CSV loaded succesfully")
            print(f"Rows: {df.shape[0]}")
            print(f"Columns: {df.shape[1]}")

            display(df.head())
        except Exception as error:
            print("Error while loading CSV file : ")
            print(error)

# connect button to function
load_button.on_click(load_csv_button_clicked)

# Display UI
app_ui = widgets.VBox([
    widgets.HTML("<h1>Mini Data Analysis Toolkit</h1>"),
    widgets.HTML("<p>v0.1 - CSV Loader</p>"),
    file_path_input,
    load_button,
    output
])

display(app_ui)
```
